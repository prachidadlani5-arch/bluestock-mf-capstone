```python
import pandas as pd
import numpy as np
```

```python
nav = pd.read_csv("../data/raw/02_nav_history.csv")
```

```python
nav.head()
```

```python
nav.columns
```

```python
nav.isnull().sum()
```

```python
nav.duplicated().sum()
```

```python
nav=nav.drop_duplicates()
```

```python
nav["date"] = pd.to_datetime(nav["date"])
```

```python
nav = nav.sort_values(
    ["amfi_code","date"]
)
```

```python
nav["nav"] = nav.groupby(
    "amfi_code"
)["nav"].ffill()
```

```python
(nav["nav"] > 0).sum()
```

```python
(nav["nav"] <= 0).sum()
```

```python
nav.to_csv(
    "../data/processed/clean_nav_history.csv",
    index=False
)
```

## Cleaning Performed on nav_history.csv

- Converted date column to datetime format.
- Sorted data by amfi_code and date.
- Checked for duplicate rows and removed them.
- Verified there were no missing NAV values.
- Applied forward-fill logic (no changes because there were no missing values).
- Validated that all NAV values are greater than 0.
- Saved cleaned dataset as clean_nav_history.csv.

```python
performance = pd.read_csv("../data/raw/07_scheme_performance.csv")
```

```python
performance.head()
```

```python
performance.columns
```

```python
performance.dtypes
```

```python
return_columns = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

for col in return_columns:
    performance[col] = pd.to_numeric(performance[col], errors="coerce")
```

```python
performance[return_columns].isnull().sum()
```

```python
performance["expense_ratio_pct"].describe()
```

```python
performance[
    (performance["expense_ratio_pct"] < 0.1) |
    (performance["expense_ratio_pct"] > 2.5)
]
```

```python
performance.duplicated().sum()
```

```python
performance = performance.drop_duplicates()
```

```python
performance.isnull().sum()
```

```python
performance.to_csv(
    "../data/processed/clean_scheme_performance.csv",
    index=False
)
```

## Cleaning Performed on scheme_performance.csv

- Validated return columns are numeric.
- Converted return columns to numeric using pd.to_numeric().
- Checked for invalid return values.
- Verified expense ratio is between 0.1% and 2.5%.
- Checked for duplicate rows and removed if present.
- Saved cleaned dataset as clean_scheme_performance.csv.
